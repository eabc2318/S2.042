# Vue Statistique de notre base de donnés

## I - Répartition des dignités au sein des tenrac

Étant donné que les dignités indiquent une hiérarchie, nous avons trouvé judicieux de la montrer en mettant en valeur le fait qu'il existe plusieurs niveaux plus ou moins accessibles.

```python
import pandas
import matplotlib.pyplot as plt

def get_occur(d):
    r = {}
    for element in d:
        if element not in r:
            r[element] = 1
        else:
            r[element] += 1
    return r

table1 = pandas.read_csv("csv_output/Dignite.csv")
table1 = table1['nom_dignite'].tolist()
resultat = get_occur(table1)

x = list(resultat.keys())
y = list(resultat.values())

plt.title("Répartition des dignité des tenracs")
plt.xlabel('Dignité')
plt.ylabel('Nombre de tenracs')

plt.bar(x,y,align='center', color='pink')
plt.show()
```

On constate que la dignité de Maitre est la plus répandue de loin, cepandant on peut aussi remarquer qu'il n'y a qu'un seul chancelier, ce qui est normal car il s'agit d'un chef.

## III - Répartition des tenracs au sein des clubs

Sachant que les clubs font partie intégrante de la vie des tenracs, il est intéréssant de regarder les plus populaires en fonction de leurs nombre de membres pour se donner une idée de leur importance.

```python
import pandas
import matplotlib.pyplot as plt

def get_occur(d):
    r = {}
    for element in d:
        if element not in r:
            r[element] = 1
        else:
            r[element] += 1
    return r

table1 = pandas.read_csv("csv_output/TenracClub.csv")
table1 = table1['id_club'].tolist()
resultat = get_occur(table1)

x = list(resultat.keys())
y = list(resultat.values())

plt.title("Répartition des tenracs par clubs")
plt.xlabel('Clubs de tenracs (ID)')
plt.ylabel('Proportion de tenracs (%)')

plt.pie(y, labels = x,            # si l'on veut "écarter" un secteur
           autopct = lambda z: str(round(z, 2)) + '%', # affichage des pourcentages dans les secteurs
           pctdistance = 0.7,                    # distance au centre pour l'affichage des pourcentages
           labeldistance = 1.2,                  # distance au centre pour l'affichage des labels
           normalize=True)                      # normalisation des données (ici déjà normalisées = somme à 1)

plt.show()
```

On constate que d'après le graphique circulaire, le club ayant pour ID le numéro 1 est de loin de club le plus populaire car il compte près de 60% de la population tenrac. En seconde position, on retrouve le club 2, puis le 3, et enfin le 4 qui est le moins populaire d'après ce diagramme

## IV - Evolution du nombre de réunions de tenracs par mois

Les réunions de tenracs sont des évenements qui constituent le centre des clubs et qui determine leur vitalité. On peut observer l'évolution du nombre de réunions chaque mois afin de savoir si les clubs ont plus ou moins d'activité au fil du temps.

```python
import pandas
import matplotlib.pyplot as plt

def get_occur(d):
    r = {}
    for element in d:
        if element not in r:
            r[element] = 1
        else:
            r[element] += 1
    return r

table1 = pandas.read_csv("csv_output/Reunion.csv")
table1 = table1['date_reunion'].tolist()
retour = []
for element in table1:
    retour.append(element[21]+element[22])
    
resultat = get_occur(retour)

x = list(resultat.keys())
y = list(resultat.values())

plt.fill_between(x, y, color="lightgreen", alpha=0.4)

plt.title("Réunions de tenracs par mois")
plt.xlabel('Mois')
plt.ylabel('Nombre de réunions')

plt.plot(x, y, "*", color="darkgreen")
plt.plot(x,y, "green")
plt.show()
```

On constate que le nombre de réunions est plutot autour de 23 par mois excepté en avril ou on tombe autour de 15 et en juin ou l'on tombe à 10 dans le mois. Il est important de remarque que le graphique saute certains mois ce qui signifie qu'il arrive fréquemment qu'il n'y ai pas de réunions durant un mois entier. On peut donc en déduire que le nombre de réunions en fonction des mois et particulièrement inégal et que les tenracs sont capable de se regrouper énormément un mois et absolument plus le suivant.

## VII - Comparaison des premières lettres des noms et prénoms des tenracs

Bien que ces données ne soient pas importantes pour les tenracs en particulier, elles démontrent un certain lien entre les lettres utilisés pour les prénom et les noms.

```python
import pandas
import matplotlib.pyplot as plt

def get_occur(d):
    r = {}
    for element in d:
        if element not in r:
            r[element] = 1
        else:
            r[element] += 1
    return r

table1 = pandas.read_csv("csv_output/Tenrac.csv")
table1 = table1['nom'].tolist()
table2 = []
for element in table1:
    table2.append(element[0])
    
table3 = pandas.read_csv("csv_output/Tenrac.csv")
table3 = table3['prenom'].tolist()
table4 = []
for element in table3:
    table4.append(element[0])

resultat1 = get_occur(table2)
resultat2 = get_occur(table4)

x1 = list(resultat1.keys())
y1 = list(resultat1.values())

x2 = list(resultat2.keys())
y2 = list(resultat2.values())

plt.title("Comparaison des premières lettres des noms (cyan) et prénoms(bleu) des tenracs")
plt.xlabel('Premières lettres')
plt.ylabel('Nombre de tenracs')

plt.bar(x1,y1,align='center', color='cyan')
plt.bar(x2,y2,color='navy',width=0.5)

plt.show()
```

On observe un rapprochement entre le nombre d'occurences des lettres P, V, M, T, O, H à la fois dans les prénoms et les noms des tenracs.
